# Zadanie 1

```julia
f(x) = x^2 - 25 * sin(x)^9
df(x) = 2x - 225 * sin(x)^8 * cos(x)
xlo = 0.1
xhi = 1.9
xs = range(xlo, xhi, length=1000)

termcrit_change(eps) = (x1, x2) -> abs(x1 - x2) < eps
termcrit_error(eps) = (x1, x2) -> abs(f(x2)) < eps

function newton(x0, maxiter, termcrit)    
    x1, x2 = Inf, x0
    i = 0
    candidates = [x2]
    
    while i < maxiter && !termcrit(x1, x2)
        x1, x2 = x2, x2 - f(x2) / df(x2)
        i += 1
        push!(candidates, x2)
    end
    candidates
end

function sieczne(x0, x1, maxiter, termcrit)
    x1, x2 = x0, x1
    i = 0
    candidates = [x2]
    
    while i < maxiter && !termcrit(x1, x2)
        x1, x2 = x2, x2 - (x2 - x1) / (f(x2) - f(x1)) * f(x2)
        i += 1
        push!(candidates, x2)
    end
    candidates
end

sieczneleft(x0, maxiter, termcrit) = sieczne(xlo, x0, maxiter, termcrit)
sieczneright(x0, maxiter, termcrit) = sieczne(xhi, x0, maxiter, termcrit)

gr()
@manipulate for method = [newton, sieczneleft, sieczneright],
                x0 = xlo:0.1:xhi,
                maxiter = 1:100,
                termcrit = [termcrit_change, termcrit_error],
                eps = range(0, 0.1, length=101)[2:end]
    
    candidates = method(x0, maxiter, termcrit(eps))
    
    iters = length(candidates) - 1
    
    plot(
        xs,
        f.(xs),
        label="f(x)",
        line=:dash,
        title="$(candidates[end])", #", $method, $iters iter., $termcrit, eps = $eps",
    )
    
    scatter!(
        candidates,
        f.(candidates),
        label="x_i",
    )
    
    plot!(
        candidates,
        f.(candidates),
        line=:arrow,
        lw=2,
        label=false,
    )
end
```

```julia
iters_newton_change = []
iters_newton_error = []
iters_sieczne_change = []
iters_sieczne_error = []

eps = 10. .^ (-1:-1:-10)

x0 = 0.5

for e = eps
    push!(iters_newton_change, length(newton(x0, 100, termcrit_change(e))) - 1)
    push!(iters_newton_error, length(newton(x0, 100, termcrit_error(e))) - 1)
    push!(iters_sieczne_change, length(sieczneleft(x0, 100, termcrit_change(e))) - 1)
    push!(iters_sieczne_error, length(sieczneleft(x0, 100, termcrit_error(e))) - 1)
end

[
    eps iters_newton_change iters_newton_error iters_sieczne_change iters_sieczne_error
]
```

# Zadanie 2
Układ równań

$
\begin{cases}
x_1^2 + x_2^2 + x_3 = 1 \\
2x_1^2 - x_2^2 - 4x_3^2 = -3 \\
x_1^2 + x_2 + x_3 = 1
\end{cases}
$

$
\begin{cases}
x_1^2 + x_2^2 + x_3 - 1 = 0 \\
2x_1^2 - x_2^2 - 4x_3^2 + 3 = 0 \\
x_1^2 + x_2 + x_3 - 1 = 0
\end{cases}
$

Jakobian

$
\left[\begin{matrix}
2x_1 & 2x_2 & 1 \\
4x_1 & -2x_2 & -8x_3 \\
2x_1 & 1 & 1
\end{matrix}\right]
$

```julia
F(x) = [
    x[1]^2 + x[2]^2 + x[3] - 1
    2x[1]^2 - x[2]^2 - 4x[3]^2 + 3
    x[1]^2 + x[2] + x[3] - 1
]

J(x) = [
    2x[1]   2x[2]  1
    4x[1]  -2x[2] -8x[3]
    2x[1]   1      1
]

solvestop_change(eps) = (x1, x2) -> sum(abs.(x1 - x2)) < eps
solvestop_error(eps) = (x1, x2) -> sum(abs.(F(x2))) < eps

function solve(x0, maxiter, termcrit)
    x1, x2 = [Inf, Inf, Inf], x0
    i = 0
    candidates = [x2]
    
    while i < maxiter && !termcrit(x1, x2)
        x1, x2 = x2, x2 - J(x2) \ F(x2)
        i += 1
        push!(candidates, x2)
    end
    candidates
end

candidates = []
goodx = []
badx = []

for x1 = -10.:10, x2 = -10.:10, x3 = -10.:10
    x0 = [x1, x2, x3]
    try
        res = solve(x0, 100, solvestop_error(0.0001))
        if length(res) == 101
            push!(badx, x0)
            continue
        end
        push!(goodx, x0)
        push!(candidates, res[end])
    catch
    end
end

candidates = [round.(sol, digits=2) for sol=candidates]
candidates = [[x === -0.0 ? 0.0 : x for x=sol] for sol=candidates]
candidates = unique(candidates)

display(candidates)

clusters = Dict()

for x0 = goodx
    sol = solve(x0, 100, solvestop_error(0.0001))[end]
    sol = round.(sol, digits=2)
    sol = [x === -0.0 ? 0.0 : x for x=sol]
    origins = get(clusters, sol, [])
    push!(origins, x0)
    clusters[sol] = origins
end

solutions = []

for x0 = candidates
    push!(solutions, solve(x0, 100, solvestop_change(0.0001))[end])
end

solutions |> display

F.(solutions) |> display

plotlyjs()
scatter3d(
    getindex.(badx, 1),
    getindex.(badx, 2),
    getindex.(badx, 3),
    markeralpha=0.5,
    markercolor=:red,
    markersize=2,
    label="Bad points",
)

for (sol, origins) = clusters
    scatter3d!(
        getindex.(origins, 1),
        getindex.(origins, 2),
        getindex.(origins, 3),
        markeralpha=0.5,
        markersize=2,
        label="$sol",
    )
end
    
plot!() |> display

;
```
